# Tratamento de dados

```python
df_DESSUST = df_DESSUST.drop([0])
df_DESSUST
```

```python
#Transformando ND em NAN
df_DESSUST = df_DESSUST.replace('ND', float('nan'))
df_DESSUST
```

```python
#Contando Nas
df_DESSUST.isnull().sum()
```

```python
#Verificando o tipo dos dados
df_DESSUST.dtypes
```

```python
#Transformando em float as colunas object
colunas = ['PIB per capita do município','  Índice da população total com atendimento de água',
           'Urbanização vias públicas', 'Escala de acesso a banda larga fixa', 'Escala de acesso a banda larga móvel',
           'Escala de acesso à banda larga fixa de alta velocidade',
          'Índice de equipamentos de tecnologia disponíveis nas escolas públicas municipais',
           'Taxa de analfabetismo',
           'Percentual de escolas municipais com acesso à internet','Números de estações rádio base', 'Computadores para uso dos alunos',
           'Taxa de homicídios']

for coluna in colunas:
    df_DESSUST[coluna] = df_DESSUST[coluna].str.replace(',', '.').astype(float)

df_DESSUST
```

```python
#Conferindo
df_DESSUST.dtypes
```

```python
#Agrupando para visualizar o valor que mais se repete(moda) nas colunas não numericas com NaN
agrupado_residuos_solidos_coleta_seletiva_de_residuos_no_municipio = df_DESSUST.groupby(['Resíduos sólidos - Coleta seletiva de resíduos no município']).size()
print(f'Agrupado da coluna: {agrupado_residuos_solidos_coleta_seletiva_de_residuos_no_municipio}')
```

```python
#Substituindo NaN pela moda
df_DESSUST['Resíduos sólidos - Coleta seletiva de resíduos no município'].fillna('NAO', inplace=True)
df_DESSUST
```

```python
#Tratando Nas restantes
#Calculando a mediana das colunas com valores númericos para substituir os NaN
mediana_Índice_da_população_total_com_atendimento_de_água = sts.median(df_DESSUST['  Índice da população total com atendimento de água'])
mediana_Urbanização_vias_públicas = sts.median(df_DESSUST['Urbanização vias públicas'])
mediana_Escala_de_acesso_à_banda_larga_fixa_de_alta_velocidade = sts.median(df_DESSUST['Escala de acesso à banda larga fixa de alta velocidade'])
mediana_Numeros_de_estações_rádio_base = sts.median(df_DESSUST['Números de estações rádio base'])
mediana_Índice_de_equipamentos_de_tecnologia_disponíveis_nas_escolas_públicas_municipais = sts.median(df_DESSUST['Índice de equipamentos de tecnologia disponíveis nas escolas públicas municipais'])
mediana_Taxa_de_analfabetismo = sts.median(df_DESSUST['Taxa de analfabetismo'])
mediana_Percentual_de_escolas_municipais_com_acesso_a_internet =  sts.median(df_DESSUST['Percentual de escolas municipais com acesso à internet'])
mediana_Computadores_para_uso_dos_alunos = sts.median(df_DESSUST['Computadores para uso dos alunos'])
mediana_Taxa_de_homicidios =  sts.median(df_DESSUST['Taxa de homicídios'])
```

```python
#Substituindo os NaN pela mediana
df_DESSUST['  Índice da população total com atendimento de água'].fillna(mediana_Índice_da_população_total_com_atendimento_de_água, inplace=True)
df_DESSUST['Urbanização vias públicas'].fillna(mediana_Urbanização_vias_públicas, inplace=True)
df_DESSUST['Escala de acesso à banda larga fixa de alta velocidade'].fillna(mediana_Escala_de_acesso_à_banda_larga_fixa_de_alta_velocidade, inplace=True)
df_DESSUST['Números de estações rádio base'].fillna(mediana_Numeros_de_estações_rádio_base, inplace=True)
df_DESSUST['Índice de equipamentos de tecnologia disponíveis nas escolas públicas municipais'].fillna(mediana_Índice_de_equipamentos_de_tecnologia_disponíveis_nas_escolas_públicas_municipais, inplace=True)
df_DESSUST['Taxa de analfabetismo'].fillna(mediana_Taxa_de_analfabetismo, inplace=True)
df_DESSUST['Percentual de escolas municipais com acesso à internet'].fillna(mediana_Percentual_de_escolas_municipais_com_acesso_a_internet, inplace=True)
df_DESSUST['Computadores para uso dos alunos'].fillna(mediana_Computadores_para_uso_dos_alunos, inplace=True)
df_DESSUST['Taxa de homicídios'].fillna(mediana_Taxa_de_homicidios, inplace=True)
df_DESSUST
```

```python
#Verificando se todas as NaN foram substituidas
df_DESSUST.isnull().sum()
```
